## Compute 'is_hum' (entry-level) and 'hum_score' (verb-level) for a given syntactic function

```python
import re
from matplotlib import pyplot as plt

def is_hum_for_synt_fn_of_entry(entry, synt_fns):
    """Returns a boolean describing whether the entry has more human (vs. non-human) possibilities for the specified syntactic functions
    in the 'FRAME' field
    """
    nb_hum_tot = 0
    nb_nhum_tot = 0

    for synt_fn in synt_fns:
        pat = rf"\??{synt_fn}(?:<\w+>)?"

        for key in entry["FRAME"].keys():
            if re.search(pat, key):
    
                selection_constraints = entry["FRAME"][key]
                
                nb_hum_tot += len(re.findall(re.compile(r"\[(\w+,\s?)*hum(,\s?\w+)*\]"), selection_constraints))
                nb_nhum_tot += len(re.findall(re.compile(r"\[(\w+,\s?)*nhum(,\s?\w+)*\]"), selection_constraints))
            
    return nb_hum_tot > nb_nhum_tot


def hum_score_for_synt_fn_of_verb(verb, synt_fns):
    """Returns a score between 0 and 1 which is the ratio of a verb's 'human entries' over total entries
    """
    nb_hum = len([entry for entry in verb["entries"] if entry[f"hum_{synt_fns[0]}"]])

    return nb_hum/len(verb["entries"])

def compute_hum_scores_for_synt_fn(data, synt_fns):
    """Stores in 'data' the 'is_hum' (at entry-level) and 'hum_score' (at verb-level) attributes for the specified syntactic functions
    """
    # if several syntactic functions considered, we use the first to name the attributes 
    synt_fn = synt_fns[0]

    for verb in data:
        for entry in data[verb]["entries"]:
            entry[f"hum_{synt_fn}"] = is_hum_for_synt_fn_of_entry(entry, synt_fns)

    for verb in data:
        data[verb][f"hum_score_{synt_fn}"] = hum_score_for_synt_fn_of_verb(data[verb], synt_fns)
```

```python
data_en = {}

for verb_fr in data:
    for entry_fr in data[verb_fr]["entries"]:
        
        verbs_en = entry_fr["TR_EN"].split(",")

        for verb_en in verbs_en:

            if verb_en == "" or not verb_en[0].isalpha():
                continue
                
            verb_en = verb_en.split(" ")[0].strip()

            sub_verbs_en = verb_en.split("/")

            for sub_verb_en in sub_verbs_en:

                if sub_verb_en == "" or not sub_verb_en[0].isalpha():
                    continue

                if sub_verb_en in data_en:
                    data_en[sub_verb_en]["entries"].append(entry_fr)
                    data_en[sub_verb_en]["num_entries"] += 1
        
                else:
                    data_en[sub_verb_en] = {
                        "entries" : [entry_fr],
                        "hum_score_subj" : None,
                        "hum_score_obj" : None,
                        "num_entries" : 1
                    }
```
